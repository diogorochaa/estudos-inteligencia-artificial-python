```python
df = pd.DataFrame({
    'Ano': [2007, 2006, 2018, 2014, 2014, 2018, 2013, 2014, 2014],
    'Preço': [80000, 25000, 30000, 25000, 75000, 10000, 60000, 90000, 80000],
    'Km': [45000, 40000, 22000, 28000, 36000, 60000, 27000, 40000, 90000],
    'Tipo': [1, 0, 1, 1, 0, 1, 0, 0, 1],
})
```

```python
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = 5  # 3,5,7 número de vizinhos
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(df[['Ano', 'Preço', 'Km']], df['Tipo'])
knn.predict([[2010, 90000, 30000]])
```

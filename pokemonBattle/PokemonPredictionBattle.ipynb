## **Preparação dos dados**

```python
#Fazendo uma cópia do dataframe e juntando os dois tipos em uma coluna só

pokemonCOPY = pokemon.copy()

pokemonCOPY['Type'] = pokemonCOPY[['Type 1', 'Type 2']].agg('/'.join, axis=1)
pokemonCOPY = pokemonCOPY.drop(['Type 1', 'Type 2'], axis=1)
pokemonCOPY.head()
```

```python
battlesAddData = battles.copy()
```

```python
# Acrescentando dados em um novo DataFrame de batalha

typeDict = dict(zip(pokemonCOPY['#'], pokemonCOPY['Type']))
hpDict = dict(zip(pokemonCOPY['#'], pokemonCOPY['HP']))
attackDict = dict(zip(pokemonCOPY['#'], pokemonCOPY['Attack']))
defenseDict = dict(zip(pokemonCOPY['#'], pokemonCOPY['Defense']))
spattackDict = dict(zip(pokemonCOPY['#'], pokemonCOPY['Sp. Atk']))
spdefenseDict = dict(zip(pokemonCOPY['#'], pokemonCOPY['Sp. Def']))
speedDict = dict(zip(pokemonCOPY['#'], pokemonCOPY['Speed']))
statsSumDict = dict(zip(pokemonCOPY['#'], pokemonCOPY['stats_sum']))
winSumDict = dict(zip(pokemonCOPY['#'], pokemonCOPY['win']))
ratioDict = dict(zip(pokemonCOPY['#'], pokemonCOPY['ratio']))



battlesAddData['First_pokemon_Type'] = battlesAddData['First_pokemon'].replace(typeDict)
battlesAddData['First_pokemon_HP'] = battlesAddData['First_pokemon'].replace(hpDict)
battlesAddData['First_pokemon_ATK'] = battlesAddData['First_pokemon'].replace(attackDict)
battlesAddData['First_pokemon_DEF'] = battlesAddData['First_pokemon'].replace(defenseDict)
battlesAddData['First_pokemon_SP_ATK'] = battlesAddData['First_pokemon'].replace(spattackDict)
battlesAddData['First_pokemon_DP_DEF'] = battlesAddData['First_pokemon'].replace(spdefenseDict)
battlesAddData['First_pokemon_Speed'] = battlesAddData['First_pokemon'].replace(speedDict)
battlesAddData['First_pokemon_STATS_SUM'] = battlesAddData['First_pokemon'].replace(statsSumDict)
battlesAddData['First_pokemon_RATIO'] = battlesAddData['First_pokemon'].replace(ratioDict)

battlesAddData['Second_pokemon_Type'] = battlesAddData['Second_pokemon'].replace(typeDict)
battlesAddData['Second_pokemon_HP'] = battlesAddData['Second_pokemon'].replace(hpDict)
battlesAddData['Second_pokemon_ATK'] = battlesAddData['Second_pokemon'].replace(attackDict)
battlesAddData['Second_pokemon_DEF'] = battlesAddData['Second_pokemon'].replace(defenseDict)
battlesAddData['Second_pokemon_SP_ATK'] = battlesAddData['Second_pokemon'].replace(spattackDict)
battlesAddData['Second_pokemon_DP_DEF'] = battlesAddData['Second_pokemon'].replace(spdefenseDict)
battlesAddData['Second_pokemon_Speed'] = battlesAddData['Second_pokemon'].replace(speedDict)
battlesAddData['Second_pokemon_STATS_SUM'] = battlesAddData['Second_pokemon'].replace(statsSumDict)
battlesAddData['Second_pokemon_RATIO'] = battlesAddData['Second_pokemon'].replace(ratioDict)
```

```python
# Função para calcular tipo/dano
def calcRelationType(combatAddData):
    r0 = 1
    firstType1 = combatAddData['First_pokemon_Type'].split('/')[0]
    firstType2 = combatAddData['First_pokemon_Type'].split('/')[1]
    secondType1 = combatAddData['Second_pokemon_Type'].split('/')[0]
    secondType2 = combatAddData['Second_pokemon_Type'].split('/')[1]
    if firstType2 != "None" and secondType2 != "None":
            r1 = dfTypeRelation[firstType1][secondType1]
            r2 = dfTypeRelation[firstType1][secondType2]
            r3 = dfTypeRelation[firstType2][secondType1]
            r4 = dfTypeRelation[firstType2][secondType2]
            r = r0 * r1 * r2 * r3 * r4
    elif firstType2 != "None" and secondType2 == "None":
            r1 = dfTypeRelation[firstType1][secondType1]
            r3 = dfTypeRelation[firstType2][secondType1]
            r = r0 * r1 * r3
    elif firstType2 == "None" and secondType2 != "None":
            r1 = dfTypeRelation[firstType1][secondType1]
            r2 = dfTypeRelation[firstType1][secondType2]
            r = r0 * r1 * r2
    elif firstType2 == "None" and secondType2 == "None":
            r1 = dfTypeRelation[firstType1][secondType1]
            r = r0 * r1
    return r

battlesAddData['Relation'] = battlesAddData.apply(lambda x: calcRelationType(x), axis=1)
```

```python
# Criando uma coluna que identifica que o primeiro pokemon ganhou
battlesAddData['First Win'] = battlesAddData.apply(lambda x: 1 if x['First_pokemon']==x['Winner'] else 0, axis=1)

# Deixando de lado as colunas que não precisamos
dropCols = ["First_pokemon", "Second_pokemon", "Winner", "First_pokemon_Type", "Second_pokemon_Type"]
combatAddData = battlesAddData.drop(dropCols, axis=1)
combatAddData.head()
```

```python
combatAddData.info()
```

```python
# Criação dos grupos de treino/teste
from sklearn.model_selection import train_test_split
X = combatAddData.drop('First Win', axis=1)
y = combatAddData['First Win']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=1)
print("X_train.shape = " + str(X_train.shape))
print("X_test.shape = " + str(X_test.shape))
print("y_train.shape = " + str(y_train.shape))
print("y_test.shape = " + str(y_test.shape))
```

## **Algoritmos**

### **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, accuracy_score
logreg = LogisticRegression(max_iter=2000)
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
precisionLogistic = precision_score(y_test, y_pred)
accuracyLogistic = accuracy_score(y_test, y_pred)

print("Precisão: ", precisionLogistic)
print("Acurácia:", accuracyLogistic)
```

### **KNN Classification**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score, accuracy_score
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
precisionKnn = precision_score(y_test, y_pred)
accuracyKnn = accuracy_score(y_test, y_pred)

print("Precisão: ", precisionKnn)
print("Acurácia:", accuracyKnn)
```

### **Gaussian Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
gaussian = GaussianNB()
gaussian.fit(X_train, y_train)
y_pred = gaussian.predict(X_test)
precisionGnb = precision_score(y_test, y_pred)
accuracyGnb = accuracy_score(y_test, y_pred)

print("Precisão: ", precisionGnb)
print("Acurácia:", accuracyGnb)
```

### **Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
decisionTree = DecisionTreeClassifier()
decisionTree.fit(X_train, y_train)
y_pred = decisionTree.predict(X_test)
precisionTree = precision_score(y_test, y_pred)
accuracyTree = accuracy_score(y_test, y_pred)

print("Precisão: ", precisionTree)
print("Acurácia:", accuracyTree)
```

### **Random Forest**


```python
from sklearn.ensemble import RandomForestClassifier
randomForest = RandomForestClassifier(n_estimators=100)
randomForest.fit(X_train, y_train)
y_pred = randomForest.predict(X_test)
precisionForest = precision_score(y_test, y_pred)
accuracyForest = accuracy_score(y_test, y_pred)

print("Precisão: ", precisionForest)
print("Acurácia:", accuracyForest)
```
